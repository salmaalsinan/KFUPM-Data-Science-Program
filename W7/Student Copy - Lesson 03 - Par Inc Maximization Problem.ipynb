# Par, Inc. Manufacturing
Par, Inc. is a small manufacturer of golf equipment and supplies whose management has decided to move into the market for medium- and high-priced golf bags. Pars distributor is enthusiastic about the new product line and has agreed to buy all the golf bags Par produces over the next three months.
After a thorough investigation of the steps involved in manufacturing a golf bag, management determined that each golf bag produced will require the following operations:
1. Cutting and dyeing the material
2. Sewing
3. Finishing (inserting umbrella holder, club separators, etc.)
4. Inspection and packaging



Pars production is constrained by a limited number of hours available in each department. After studying departmental workload projections, the director of manufacturing estimates that:
- 630 hours for cutting and dyeing,
- 600 hours for sewing,
- 708 hours for finishing, and
- 135 hours for inspection and packaging

will be available for the production of golf bags during the next three months.

## Problem Modeling
Problem formulation, or modeling, is the process of translating the verbal statement of a
problem into a mathematical statement. Formulating models is an art that can be mastered
only with practice and experience. Even though every problem has some unique features,
most problems also have common features. As a result, some general guidelines for optimization model formulation can be helpful, especially for beginners. We will illustrate these
general guidelines by developing a mathematical model for Par, Inc.

### Describe the Objective
The objective is to maximize the total contribution to profit.

### Describe each constraint
Four constraints relate to the number of hours of manufacturing time available; they restrict the number of standard bags and the number of deluxe bags that can be produced.
- **Constraint 1:** The number of hours of cutting and dyeing time used must be less than or equal to the number of hours of cutting and dyeing time available.
- **Constraint 2:** The number of hours of sewing time used must be less than or equal to the number of hours of sewing time available.
- **Constraint 3:** The number of hours of finishing time used must be less than or equal to the number of hours of finishing time available.
- **Constraint 4:** The number of hours of inspection and packaging time used must be less than or equal to the number of hours of inspection and packaging time available.

### Define the decision variables
The controllable inputs for Par, Inc. are:
1. the number of standard bags produced and
2. the number of deluxe bags produced.

Let:
<p style="text-align: center;">
$S$ = number of standard bags, and<br/>
$D$ = number of deluxe bags
</p>

In optimization terminology, $S$ and $D$ are referred to as the **decision variables.**

### Write the objective in terms of the decision variables
Pars profit contribution comes from two sources:
1. the profit contribution made by producing $S$ standard bags and
2. the profit contribution made by producing $D$ deluxe bags.

If Par makes \\$10 for every standard bag, the company will make \\$$10S$ if $S$ standard bags are produced. Also, if Par makes \\$9 for every deluxe bag, the company will make \\$$9D$ if $D$ deluxe bags are produced. Thus, we have

_Total profit contribution = $10S + 9D$_

Because the objective maximize total profit contribution is a function of the decision variables $S$ and $D$, we refer to $10S + 9D$ as the **objective function**

<p style="text-align: center;">$\max{(10S+9D)}$</p>

#### Constraint 1: Cutting and Dyeing
Total hours of cutting and dyeing time used = $\frac{7}{10}S + 1D$

The director of manufacturing stated that Par has at most 630 hours of cutting and dyeing time available

<p style="text-align: center;">$\frac{7}{10}S + 1D \leq 630$</p>

#### Constraint 2: Sewing
<p style="text-align: center;">$\frac{1}{2}S + \frac{5}{6}D \leq 600$</p>

#### Constraint 3: Finishing
<p style="text-align: center;">$1S + \frac{2}{3}D \leq 708$</p>

#### Constraint 4: Inspection
<p style="text-align: center;">$\frac{1}{10}S + \frac{1}{4}D \leq 135$</p>

#### Nonnegative constraints
Par, Inc. cannot produce negative number of standard or deluxe bags.
<p style="text-align: center;">$S \geq 0$ and $ D \geq 0$</p>

This mathematical model of the Par, Inc. problem is a **linear programming model**, or **linear program**, because the objective function and all constraint functions (the left-hand sides of the constraint inequalities) are linear functions of the decision variables.

<blockquote>Mathematical functions in which each variable appears in a separate term and is raised
    to the first power are called <b>linear functions</b>. </blockquote>
    
### Solving the Par. Inc. Problem
A feasible solution is a setting of the decision variables that satisfies all of the constraints of
the problem. The optimal solution also must have an objective function value as good as any other feasible solution.

#### The Geometry of the Par, Inc. Problem

If the constraints are inequalities, the constraint cuts the space into two, with the line and the area on one side of the line being the space that satisfies that constraint. These subregions are called _half spaces_. The _intersection_ of these half spaces makes up the feasible region.

````python
import numpy as np
import matplotlib.pyplot as plt

s = np.linspace(0, 1400, 1000)
d1 = 630 - (7./10.)*s
d2 = (6./5.)*(600 - s/2.)
d3 = (3./2.)*(708 - s)
d4 = 4.*(135 - s/10.)
d5_y_0 = 1+s*0


plt.plot(s, d1, 'b--', label=r'$\frac{7}{10}S + 1D \leq 630$')
plt.plot(s, d2, 'k--', label=r'$\frac{1}{2}S + \frac{5}{6}D \leq 600$$')
plt.plot(s, d3, label=r'$1S + \frac{2}{3}D \leq 708$')
plt.plot(s, d4, label=r'$\frac{1}{10}S + \frac{1}{4}D \leq 135$')

plt.ylim(0, 1200)

plt.show()
````

````python
d_min = np.minimum(d1, d2)
d_min = np.minimum(d_min, d3)
d_min = np.minimum(d_min, d4)

plt.plot(s, d1, 'b--', label=r'$\frac{7}{10}S + 1D \leq 630$')
plt.plot(s, d2, 'k--', label=r'$\frac{1}{2}S + \frac{5}{6}D \leq 600$$')
plt.plot(s, d3, label=r'$1S + \frac{2}{3}D \leq 708$')
plt.plot(s, d4, label=r'$\frac{1}{10}S + \frac{1}{4}D \leq 135$')

plt.plot(s, d_min, 'r')

plt.ylim(0, 1200)

plt.fill_between(s, d_min, d5_y_0, facecolor='red', interpolate=True, alpha=0.3)

plt.show()
````

to solve a linear optimization problem we only have to search over the **extreme points** of the feasible region to find an optimal solution.








It turns out that this approach of investigating only extreme points works well and generalizes for larger problems. 

## Using PyMathProg Package for Linear Programming

use ``` pip install pymprog ``` command to instal the PyMathProg package in your invironment.

````python
import pymprog as pm
pm.begin('ParInc')
pm.verbose(True)
x, y = pm.var('x, y') # variables
pm.maximize(10 * x + 9 * y, 'profit')
(7./10.)*x + y <= 630 # Cutting and Dyeing
(1./2.)*x + (5./6.)*y <= 600 # Sewing
x + (2./3.) * y <= 708 # Finishing
(1./10.) * x + (1./4.)*y <= 135 # Inspection
#x >= 0 # exclude non-negative
#y >= 0 #
pm.solve()
print('Max profit:',pm.vobj())
pm.sensitivity()
pm.end()
````

#### The Dual Value (Shadow Price)
For binding constratins, indicated by the constraints **not** preceeded by (\*) in the output table above, the Dual Value a.k.a Shadow Price show the change in the optimal objective function value if the right-hand side of that constraint is increased by one.

<p style='color:red;'>Verify this by changing the constraint's right hand value and running the optimization again:</p>

The **slack** value for each less-than-or-equal-to constraint indicates the difference between the left-hand and right-hand values for a constraint. Of course, by definition, binding constraints have zero slack.

Consider for example the sewing department constraint. By adding a nonnegative slack variable, we can make the constraint equality:



The (\*) preceeding R2 and R4 in the output of the previous code snippet indicates that these are non-binding constraints. 



## Using PuLP Library in Python for Linear Programming

use ``` pip install pulp ``` command to install the PuLP library package in your environment.

````python
from pulp import *
prob = LpProblem("ParInc", LpMaximize)
s = LpVariable("s", lowBound=0, cat='Integer')
d = LpVariable("d", lowBound=0, cat='Integer')
prob += 10*s + 9*d
prob += (7./10.)*s + d <= 630
prob += (1./2.)*s + (5./6.)*d <= 600
prob += s + (2./3.) * d <= 708
prob += (1./10.) * s + (1./4.)*d <= 135
````

````python
status = prob.solve()
LpStatus[status]
````

````python
print ("We can produce", value(s),'\n',"standard bags and", value(d),'\n' ,"deluxe bags with maximum profit of $", value(prob.objective))
````

````python
for v in prob.variables():
    print (v.name, "=", v.varValue, "Item Cost = $", v.dj)
print ("objective=", value(prob.objective))
print ("\nSensitivity Analysis\nConstraint\t\t\t\tShadow Price\tSlack")
for name, c in prob.constraints.items():
    print( name, ":", c, "\t\t\t|", c.pi, "\t\t|", c.slack)
````

## Alternative Optimal Solutions
Now let us consider the special case in which the optimal objective function contour line coincides with one of the binding constraint lines on the boundary of the feasible region. We will see that this situation can lead to the case of **alternative optimal solutions**; in such cases, more than one solution provides the optimal value for the objective function.

Let us assume that the profit for the standard golf bag (S) has been decreased to \\$6.30. The revised objective function becomes $6.3 S + 9D$

<p style='color:red;'>Use the optimization code written using PyMProg library above to solve this optimization problem:</p>

````python
import pymprog as pm
pm.begin('ParInc')
pm.verbose(True)
x, y = pm.var('x, y') # variables
pm.maximize(6.3 * x + 9 * y, 'profit')
(7./10.)*x + y <= 630 # Cutting and Dyeing
(1./2.)*x + (5./6.)*y <= 600 # Sewing
x + (2./3.) * y <= 708 # Finishing
(1./10.) * x + (1./4.)*y <= 135 # Inspection
#x >= 0 # exclude non-negative
#y >= 0 #
pm.solve()
print('Max profit:',pm.vobj())
pm.sensitivity()
pm.end()
````

## Unbounded

The solution to a maximization linear programming problem is **unbounded** if the value of the solution may be made infinitely large without violating any of the constraints; for a minimization problem, the solution is unbounded if the value may be made infinitely small.
