## Exercise 1

```python
import pandas as pd

df = pd.read_csv('employees.csv', index_col='Name')

```

```python
mapping = {True: 'T', False: 'F', None: 'NA'} 

df['Senior Management'] = df['Senior Management'].replace(mapping)
```

```python
df
```

```python
# “Training” such that the employees having “C” or “PF” positions are eligible to attend the training.

mymap = {"C": True, "PF": True}
df['Training'] = df['Position'].map(mymap, '')
```

```python
df
```

```python
salaries = df['Salary']
labels = ['low', 'average', 'high']
s_cat = pd.cut(salaries, 3, labels=labels)
```

```python
s_cat
```

```python
pd.value_counts(s_cat)
```

```python
qcat = pd.qcut(df['Salary'], 3)
```

```python
qcat
```

```python
pd.value_counts(qcat)
```
